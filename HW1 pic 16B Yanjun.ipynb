## Part 3: Write a query function (10 pts)

Write a function called `query_climate()` which accepts five arguments:

`df`, the file name for the dataframe

`country`, a string giving the name of a country for which data should be returned.

`year_begin` and `year_end`, two integers giving the earliest and latest years for which should be returned.

`month`, an integer giving the month of the year for which should be returned.

The return value of `query_climate()` is a Pandas dataframe of temperature readings for the specified country, in the specified date range, in the specified month of the year. This dataframe should have the following columns, **in this order**:

`NAME`: The station name.

`LATITUDE`: The latitude of the station.

`LONGITUDE`: The longitude of the station.

`Country`: The name of the country in which the station is located.

`Year`: The year in which the reading was taken.

`Month`: The month in which the reading was taken.

`Temp`: The temperature at the specified station during the specified year and month.


**Grading is based on unseen test examples. A test example is given below.**

```python
def query_climate(df, country, year_begin, year_end, month):
    """
    Function to query climate data of temperature readings for the specified country, in the specified date range, in the specified month of the year.
    """
    if isinstance(country, list):
        # if country is a list
        df = df[df['Country'].isin(country)]
    else:
        # if country is a single string
        df = df[df['Country'] == country]
    df = df.query("Year >= @year_begin & Year <= @year_end")
    # locate the data of specific month
    returned_df = df.loc[:,['NAME', "LATITUDE","LONGITUDE", "Country","Year"]]
    returned_df['Month'] = month
    returned_df['Temp'] = df.loc[:,['VALUE'+str(month)]]
    returned_df = returned_df.sort_values(['NAME', "Year"])
    return returned_df
```

## Part 4: Global warming ( 10 pts )

In this part, the goal is to create a bar plot which is used to compare average temperatures of decadeds 1901-1910 and 2011-2020.

The bar plot looks like the following:



In above plot, I select three different countries and compute the average temperatures of year range 1901-1910 and 2011-2020 in January. Then, I use (two) barplot(s) to visualize the average temperature.

**You should do more than my sample plot:**

Select 4 different countries and 6 different months, then compute two average temperatures for each country in each certain month. Then visualize 6 different plots (included in a single plot using `subplot` command). Please add necessary labels, titles, and legends to make your plot clear.

You should write a function or class to simplify your code if necessary.

What is your conclusion?

```python
countries = ['India', 'United States', 'Japan', 'Algeria'] # 4 countries
months = [1, 3, 5, 7, 9, 11] # 6 months

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
def compute_mean_temp(df, countries, year_begin, year_end, month):
    """
    Function to compute the average temperature for the specified country, in the specified date range, in the specified month of the year.
    """
    means = []
    for country in countries:
      means.append(query_climate(df, country, year_begin, year_end, month)['Temp'].mean())

    return means

def global_warming(df = df, countries = countries, months = months):
    """
    Function to plot global warming
    """
    fig, ax = plt.subplots(2,3, figsize=(15,10))
    ax = ax.flatten()
    palette = {"1901-1910": "pink", "2011-2020": "red"}
    for i in range(len(months)):
      # compute the mean temperature for each country at each decade and current month

      data = pd.DataFrame({'Country': countries+countries,
                          "decade": len(countries)*['2011-2020']+['1901-1910'] * len(countries) ,
                          'Temp': compute_mean_temp(df, countries, 2011, 2020, months[i]) + compute_mean_temp(df, countries, 1901, 1910, months[i])})
      sns.barplot(data, x='Country', y='Temp', hue = 'decade', ax=ax[i], dodge=False,palette = palette )

      ax[i].set_title(f'Average temperature of {pd.to_datetime(months[i], format="%m").strftime("%B")}')

    plt.tight_layout()
    plt.show()
global_warming(df,countries, months)
```

```python
df1 = df.set_index(keys =["ID",'LATITUDE','LONGITUDE','NAME', 'Country', 'Year'])
df1=df1.stack().reset_index()

# Add Month, date column
df1.rename(columns = {'level_6':'Month', 0:'Temp'}, inplace = True)
df1['Month'] = df1['Month'].str[5:].astype(int)
df1['Date'] = df1['Year'].astype(str) + '-' + df1['Month'].astype(str)
df1['Date'] = pd.to_datetime(df1['Date'])
df1.head()
df1
```

```python
old_data = df1.query('Year < 2000')
new_data = df1.query('Year >= 2000')
old_averages = old_data.groupby(['Country', 'Date'])[['Temp']].mean()
new_averages = new_data.groupby(['Country', 'Date'])[['Temp']].mean()
old_averages
```

# Figure 1:
This figure has 2 plots, the first plot shows the temperature change over 1901 - 1911 for 8 Asian countries, the second shows that of 2011 - 2021. As the plots show, over the years, the hotter countries has gotten hotter and the colder are colder. This might indicate the climate changes to be more extreme around the world.

```python
old_averages = old_averages.reset_index()
new_averages = new_averages.reset_index()

asian_countries = [
    "China",
    "India",
    "Indonesia",
    "Japan",
    "Vietnam",
    "Thailand",
    "Myanmar",
    "South Korea",
    "Iraq"
]
asia1 = old_averages[old_averages['Country'].isin(asian_countries)]
asia2 = new_averages[new_averages['Country'].isin(asian_countries)]

fig, ax = plt.subplots(2,1, figsize=(20,10),sharey=True)
ax = ax.flatten()
sns.lineplot(data = asia1, x = 'Date', y = 'Temp', hue = 'Country',ax =ax[0],legend = False)
sns.lineplot(data = asia2, x = 'Date', y = 'Temp', hue = 'Country',ax =ax[1])
plt.legend(bbox_to_anchor=(1.05, 1),loc=2)
plt.title('Temperature change of Asian Countries')
plt.tight_layout()
plt.show()
```

# Figure 2
In part 4, the result of 4 countries suggest there might be a global trend of warming. Thus for this  figure I chose to study the the temperature changes over 3 continents: Asia, America, and Africa. As the result shows, more the the countries across the 3 continents has increased their average temperatures, and this suppose the hypothesis of global warming.

```python
asia1['decade']= '1901-1910'
asia2['decade'] = '2011-2020'
asia = pd.concat([asia1, asia2])

american_countries = [
    "United States",
    "Canada",
    "Brazil",
    "Mexico",
    "Argentina",
    "Cuba",
    "Chile",
    "Peru"
]
american1 = old_averages[old_averages['Country'].isin(american_countries)]
american2 = new_averages[new_averages['Country'].isin(american_countries)]

american1['decade'] = '1901-1910'
american2['decade'] = '2011-2020'
american= pd.concat([american1, american2])

african_countries = [
    "Nigeria",
    "Egypt",
    "South Africa",
    "Kenya",
    "Ethiopia",
    "Ghana",
    "Morocco",
    "Tanzania"
]
african1 = old_averages[old_averages['Country'].isin(african_countries)]
african2 = new_averages[new_averages['Country'].isin(african_countries)]
african1['decade'] = '1901-1910'
african2['decade'] = '2011-2020'
african = pd.concat([african1, african2])

fig, ax = plt.subplots(3,1, figsize=(10,10), sharey = True)
ax.flatten()
sns.barplot(data = asia, x = 'Country', y = 'Temp',hue = 'decade', palette = 'Set1',estimator ='mean', ax = ax[0])
plt.legend(bbox_to_anchor=(1.05, 1),loc=2)
# plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1))
sns.barplot(data = american, x = 'Country', y = 'Temp',hue = 'decade', palette = 'Set1',estimator ='mean', ax = ax[1])
plt.legend(bbox_to_anchor=(1.05, 1),loc=2)
# plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1))

sns.barplot(data = african, x = 'Country', y = 'Temp',hue = 'decade', palette = 'Set1',estimator ='mean', ax = ax[2])
plt.legend(bbox_to_anchor=(1.05, 1),loc=2)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.tight_layout()
ax[0].set(xlabel ='Asian Countries')
ax[1].set(xlabel =' American Countries')
ax[2].set(xlabel =' African Countries')
plt.show()
```

```python
asia['Continent'] = 'Asia'
american['Continent'] = 'America'
african['Continent'] = 'Africa'
whole = pd.concat([asia, american, african])
whole
```

```python
sns.catplot(
    whole, kind="swarm",
    x="Country", y="Temp", col="Continent", hue="decade",
    height=2.5, aspect=5, alpha = 0.6,
    palette = 'Set2',
    col_wrap=1
)
plt.tight_layout()
plt.show()
```
